```python
import pandas as pd

# Load the CSV files
movies_df = pd.read_csv("tmdb_5000_movies.csv")
credits_df = pd.read_csv("tmdb_5000_credits.csv")

# Merge the datasets on title
movies = movies_df.merge(credits_df, on='title')

# Show shape and a preview of the merged data
print("Shape of merged dataset:", movies.shape)
movies.head()
```

```python
# Keep only useful columns
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]

# Preview
movies.head()
```

```python
import ast

# Function to extract name from a list of dictionaries
def extract_names(obj):
    names = []
    try:
        for item in ast.literal_eval(obj):
            names.append(item['name'])
    except:
        pass
    return names

# Clean each feature
movies['genres'] = movies['genres'].apply(extract_names)
movies['keywords'] = movies['keywords'].apply(extract_names)

# For cast, get only top 3 actors
def extract_top_cast(obj):
    names = []
    try:
        for item in ast.literal_eval(obj)[:3]:
            names.append(item['name'])
    except:
        pass
    return names

movies['cast'] = movies['cast'].apply(extract_top_cast)

# For crew, only get the director's name
def extract_director(obj):
    try:
        for item in ast.literal_eval(obj):
            if item['job'] == 'Director':
                return [item['name']]
    except:
        pass
    return []

movies['crew'] = movies['crew'].apply(extract_director)
```

```python
movies[['title', 'genres', 'keywords', 'cast', 'crew']].head()
```

```python
# Convert overview to list so it can be added with other lists
movies['overview'] = movies['overview'].apply(lambda x: [] if pd.isna(x) else [x])

# Now combine all features into one 'tags' column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

# Convert the list into a single string
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))

# Lowercase everything for uniformity
movies['tags'] = movies['tags'].apply(lambda x: x.lower())

# Preview the cleaned tags
movies[['movie_id', 'title', 'tags']].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Create a CountVectorizer with max features and stopword removal
cv = CountVectorizer(max_features=5000, stop_words='english')

# Transform 'tags' into vectors
vectors = cv.fit_transform(movies['tags']).toarray()

# Compute cosine similarity matrix
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie = movie.lower()
    if movie not in movies['title'].str.lower().values:
        print(f"'{movie}' not found in the movie list.")
        return

    index = movies[movies['title'].str.lower() == movie].index[0]
    distances = list(enumerate(similarity[index]))
    movies_list = sorted(distances, key=lambda x: x[1], reverse=True)[1:6]

    print(f"\nðŸŽ¬ Because you liked '{movies.iloc[index].title}', you may also like:\n")
    for i in movies_list:
        print("âž¤", movies.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
